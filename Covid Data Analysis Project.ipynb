```python
# import covid vaccination dataset

vaccine_df = pd.read_csv("E:\RAVI\DataSets\Covid dataset\covid_vaccine_statewise.csv")
```

```python
vaccine_df.head()
```

```python
covid_df.drop(["Sno","ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df.head()
```

```python
# Changing Date Format

covid_df["Date"] = pd.to_datetime(covid_df["Date"], format = "%Y-%m-%d")
```

```python
covid_df.head()
```

```python
#Active cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] +  covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
#Cleaning Unwanted Data

statewise = pd.pivot_table(covid_df,values = ["Confirmed","Deaths","Cured"],
                          index = "State/UnionTerritory", aggfunc = max)
```

```python
# Finding Statewise Recovery Rate

statewise["Recovery_Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
# Mortality Rate Statewise

statewise["Mortality_Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
# Statewise Covid Data

statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap="cubehelix")
```

```python
# Top 10 Active Cases in India

top_10_active_cases =covid_df.groupby(by = 'State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States most active cases in India")
ax = sns.barplot(data= top_10_active_cases.iloc[:10], y="Active_Cases",x="State/UnionTerritory",linewidth = 2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Top 10 states highest deaths

top_10_deaths =covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 States deaths in India")
ax = sns.barplot(data= top_10_deaths.iloc[:12], y="Deaths",x="State/UnionTerritory",linewidth = 2, edgecolor = 'red')
plt.xlabel("Deaths")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Growth Trend

fig = plt.figure(figsize = (12,6))

ax = sns.lineplot(data = covid_df[covid_df["State/UnionTerritory"].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x = 'Date', y= 'Active_Cases', hue = "State/UnionTerritory")

ax.set_title("Top 5 affected states in India", size = 16)
```

```python
# Remove rows where state = India

vaccine = vaccine_df[vaccine_df.State!="India"]
```

```python
vaccine
```

```python
# Rename Total Vaccinated Column to Total

vaccine =vaccine.rename(columns = {"Total Individuals Vaccinated":"Total"})
```

```python
# Most vaccinated states

max_vac = vaccine.groupby("State")['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize=(7,5))
plt.title("Top 5 Vaccinated States in India")
ax = sns.barplot(data= max_vac.iloc[:10], y=max_vac.Total ,x= max_vac.index ,linewidth = 2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

```python
# Least vaccinated states

min_vac = vaccine.groupby("State")['Total'].sum().to_frame('Total')
min_vac = min_vac.sort_values('Total',ascending = True)[:5]
min_vac
```

```python
fig = plt.figure(figsize=(15,5))
plt.title("Least 5 Vaccinated States in India")
ax = sns.barplot(data= min_vac.iloc[:10], y=min_vac.Total ,x= min_vac.index ,linewidth = 2, edgecolor = 'red')
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
